```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import skfuzzy
import numpy as np
```

# read data and normalize

```python
df = pd.read_csv('data1.csv')
```

```python
plt.scatter(df['X'], df['Y'], c=df['Class'])
plt.title('before normalization')
plt.show()
```

```python
features = df[['X', 'Y']]
scaler = StandardScaler()
df[['X', 'Y']] = scaler.fit_transform(features)
```

```python
plt.scatter(df['X'], df['Y'], c=df['Class'])
plt.title('after normalizatoin')
plt.show()
```

# clustring with fcm

```python
colors = ['red', 'green', 'blue', 'yellow', 'black', 'gray', 'orange', 'purple', 'cyan', 'magenta']
fpcs = []
for num_clusters in range(2, 11):
    cntr, u, u0, d, jm, p, fpc = skfuzzy.cluster.cmeans(df[['X', 'Y']].T, num_clusters, 2, error=0.005, maxiter=1000, init=None)
    cluster_membership = np.argmax(u, axis=0)
    fpcs.append(fpc)
    fig, ax = plt.subplots()
    
    for i in range(num_clusters):
        ax.plot(df['X'][cluster_membership == i], df['Y'][cluster_membership == i], 'o', markersize=3, label=f'Cluster {i + 1}', color=colors[i])
        ax.plot(cntr[i][0], cntr[i][1], 'X', markersize=10, markeredgewidth=1, label='Cluster Center', color=colors[i])
    ax.set_title(f'c = {num_clusters}')
    # ax.legend()
    plt.show()
```

# Best clustring

```python
plt.plot(range(2, 11), fpcs)
```

```python
best_c = np.argmax(fpcs) + 2
cntr, u, u0, d, jm, p, fpc = skfuzzy.cluster.cmeans(df[['X', 'Y']].T, best_c, 2, error=0.005, maxiter=1000, init=None)
cluster_membership = np.argmax(u, axis=0)
fig, ax = plt.subplots()
for i in range(best_c):
    ax.plot(df['X'][cluster_membership == i], df['Y'][cluster_membership == i], 'o', markersize=3, label=f'Cluster {i + 1}', color=colors[i])
    ax.plot(cntr[i][0], cntr[i][1], 'X', markersize=10, markeredgewidth=1, label='Cluster Center', color=colors[i])
ax.set_title(f'best clustring with c = {best_c} and fpc = {fpc}')
plt.show()
```
